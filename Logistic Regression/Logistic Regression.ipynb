## Logistic Regression

Logistic Regression is a statistical and machine learning algorithm used for binary classification problems (i.e., where the target variable has only two possible outcomes, like "yes/no," "spam/not spam," or "pass/fail"). Despite its name, logistic regression is actually a classification algorithm, not a regression algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = np.array([16,24,31,29,40,33,18,36,15,37,19,21,19,31,23,20,34,22,32])
job = np.array([0,0,1,0,1,0,0,1,0,1,0,0,1,1,1,0,1,0,1])
```

```python
plt.scatter(age,job, color = 'red')
plt.xlabel('Age')
plt.ylabel('Job')
plt.title('Age vs Job')
plt.show()
```
